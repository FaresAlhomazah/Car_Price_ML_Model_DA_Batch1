```python
def detect_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return column[(column < lower_bound) | (column > upper_bound)]
```

```python
for outlier in outlier_columns_check:
  outliers = detect_outliers(car_pricing[outlier])
  print(f"\n The outlyer in {outlier} ")
  print(outliers)


```

```python
# Adjust bins for enginesize categorical
car_pricing["enginesize_bins"] = pd.cut(car_pricing["enginesize"],
                                        bins=[0., 100, 150,200, np.inf],
                                        labels=[1, 2, 3,4])
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(car_pricing, car_pricing["enginesize_bins"]):
    strat_train_set = car_pricing.loc[train_index]
    strat_test_set = car_pricing.loc[test_index]
```

#Prepare The Data for ML Model

```python
car_pricing_features = strat_train_set.drop("price", axis=1)
car_pricing_labels =strat_train_set


car_pricing_features.shape , car_pricing_labels.shape
```

```python
car_pricing_features_num = car_pricing_features.drop(['CarName', 'fueltype', 'aspiration', 'doornumber'
, 'carbody', 'drivewheel',  'enginelocation', 'enginetype', 'cylindernumber', 'fuelsystem'], axis=1)
car_pricing_features_num.shape
```

```python
car_pricing_features.isnull().sum()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder
```

```python
numerical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median'))
    ,('standard_scalar' , StandardScaler())
])
#car_pricing_features_num_tr = numerical_pipeline.fit_transform(car_pricing_features_num)
```

```python
from sklearn.compose import ColumnTransformer
num_attribs = list(car_pricing_features_num)
cat_attribs = ['CarName', 'fueltype', 'aspiration', 'doornumber'
, 'carbody', 'drivewheel',  'enginelocation', 'enginetype', 'cylindernumber', 'fuelsystem']
main_pipeline = ColumnTransformer([
    ('numerical', numerical_pipeline, num_attribs )
    ,('categorical', OrdinalEncoder(), cat_attribs)

])
prepared_car_pricing_train = main_pipeline.fit_transform(car_pricing_features)
```

#Train and Test the Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
```

#Fine Tune Forest Regressor and XGB Regressor Models

#Testing Forest Regressor with Fine Tune Algorithm

```python
from sklearn.model_selection import GridSearchCV
grid_parameter =[{ 'bootstrap': [True], 'n_estimators' :list(range(10,31)), 'max_features': list(range(2,16))},
                 {'bootstrap': [False], 'n_estimators': list(range(10,31)), 'max_features' : list(range(2,16))}
                 ]

forest_reg_model2 = RandomForestRegressor()
grid_searching_process = GridSearchCV(forest_reg_model2, grid_parameter, cv=5,
                                      scoring='neg_mean_squared_error', return_train_score= True)
grid_searching_process.fit(prepared_car_pricing_train, car_pricing_labels)
```

```python
grid_searching_process.best_params_
```

```python
grid_searching_process.best_estimator_
```

```python
print(f"The mean squared error of fine tuning Random Forest Regressor model is: {np.sqrt(-grid_searching_process.best_score_)}")
```

#Testing XGB Regressor with Fine Tune Algorithm

```python
param_grid = {
    'n_estimators': list(range(10, 40)),  # Starting from 10 to 40
    'max_depth': [6],
    'learning_rate':[ 0.2],
    'subsample':[0.7],
    'colsample_bytree': [0.9]
}
#####################################################
#The best choice for param_grid is: {'colsample_bytree': 0.9,
#'learning_rate': 0.2,
#'max_depth': 6,
#'n_estimators': 38
#'subsample': 0.7}
#######################################################################
xgb_reg_model2 = xgb.XGBRegressor(objective='reg:squarederror')


xgb_grid_searching_process = GridSearchCV(estimator=xgb_reg_model2, param_grid=param_grid, cv=5
                          , scoring='neg_mean_squared_error', return_train_score=True)


xgb_grid_searching_process.fit(prepared_car_pricing_train, car_pricing_labels)
```

```python
xgb_grid_searching_process.best_params_
```

```python
xgb_grid_searching_process.best_estimator_
```

```python
np.sqrt(-xgb_grid_searching_process.best_score_)
```

We can find that the xgb_reg_model2 is the best model which is fined tuning by GridSearchCV algorithm

#Evaluate and Test the best models

```python
#Initialize the best chosen model
the_best_model = xgb_grid_searching_process.best_estimator_

#preparing the testing data
X_test = strat_test_set.drop(["price"], axis=1)
Y_test = strat_test_set["price"].copy()
X_test_prepared = main_pipeline.fit_transform(X_test)

#Test the model with testing data
final_model_predictions = the_best_model.predict(X_test_prepared)

final_mse = mean_squared_error(Y_test, final_model_predictions)
final_rmse = np.sqrt(final_mse)
final_rmse
```

```python
#This helps us understand the range within which the true RMSE of the model likely lies.
from scipy import stats

confidence = 0.95
squared_errors = (final_model_predictions - Y_test) ** 2
np.sqrt(stats.t.interval(confidence, len(squared_errors) - 1,
                         loc=squared_errors.mean(),
                         scale=stats.sem(squared_errors)))
```

#Evaluating the Accuracy of the Model

```python
price_min = car_pricing['price'].min()
price_max =car_pricing['price'].max()
price_range = price_max - price_min

normalized_rmse = final_rmse / price_range

accuracy_percentage = (1 - normalized_rmse) * 100

print(f"Accuracy-like percentage: {accuracy_percentage:.2f}%")
```
